```python
from IPython.display import HTML
import difflib


def compare_strings_html(s1, s2):
    
    diff = difflib.ndiff(s1.split(), s2.split())

    table = "<table>"

    for word in diff:
        if word[0] == '-':
            table += "<tr><td style='color: red;'>{}</td><td></td></tr>".format(word)  # Red color for deletions
        elif word[0] == '+':
            table += "<tr><td></td><td style='color: green;'>{}</td></tr>".format(word)  # Green color for additions
        else:
            table += "<tr><td>{}</td><td>{}</td></tr>".format(word, word)

    table += "</table>"

    return table

s1 = "This is line one"
s2 = "That was line one"
display(HTML(compare_strings_html(s1, s2)))
```
